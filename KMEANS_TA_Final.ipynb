```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

# Copy the cleaned dataset
df1 = df_cleaned.copy()

# Drop irrelevant columns
df1 = df1.drop(columns=["DISTRICTNAME", "UNITNAME", "Year"])

# Identify categorical columns for encoding
categorical_cols = ["Severity", "Accident_Classification", "Main_Cause", "Weather", 
                    "Road_Condition", "Surface_Condition", "Road_Character", "Road_Type"]

# Apply One-Hot Encoding
encoder = OneHotEncoder(sparse=False)  # sparse=False to get a dense matrix
encoded_features = encoder.fit_transform(df1[categorical_cols])

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_cols))

# Combine with original dataset (excluding original categorical columns)
df1 = pd.concat([df1.drop(columns=categorical_cols), encoded_df], axis=1)

# Apply StandardScaler to Latitude & Longitude
scaler = StandardScaler()
df1[["Latitude", "Longitude"]] = scaler.fit_transform(df1[["Latitude", "Longitude"]])

# Display updated dataset info
df1.info(), df1.head()
```

```python
# Fill missing values for numerical columns with the mean
df1["Latitude"].fillna(df1["Latitude"].mean(), inplace=True)
df1["Longitude"].fillna(df1["Longitude"].mean(), inplace=True)
df1["Noofvehicle_involved"].fillna(df1["Noofvehicle_involved"].mean(), inplace=True)

# Fill missing values for categorical columns with the most frequent value (mode)
df1["Severity_ACHARI"].fillna(df1["Severity_ACHARI"].mode()[0], inplace=True)







# For the one-hot encoded columns, if missing, you can fill with 0 or the mode (i.e., the most frequent category)
encoded_cols = encoder.get_feature_names_out(categorical_cols)
df1[encoded_cols] = df1[encoded_cols].fillna(0)  # Assuming 0 for missing values in encoded columns

# Check for remaining missing values
df1.isnull().sum()

# Display updated dataset info
df1.info()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import folium

# Assuming df1 is your cleaned and preprocessed DataFrame

# Select relevant features for clustering (including only essential ones)
features = df1[['Latitude', 'Longitude', 'Noofvehicle_involved',
                'Severity_Fatal', 'Severity_Grievous Injury', 
                'Weather_Clear', 'Weather_Cloudy', 'Weather_Heavy Rain', 
                'Weather_Light Rain', 'Road_Type_NH', 'Road_Type_Residential Street']]

# Scale the features using StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Apply KMeans clustering (choose number of clusters, e.g., 3)
kmeans = KMeans(n_clusters=3, random_state=42)
df1['Cluster'] = kmeans.fit_predict(features_scaled)

# Check the first few rows of the data with cluster assignments
print(df1[['Latitude', 'Longitude', 'Cluster']].head())

# Visualize the clusters (just using Latitude and Longitude for visualization)
plt.figure(figsize=(8, 6))
plt.scatter(df1['Longitude'], df1['Latitude'], c=df1['Cluster'], cmap='viridis', marker='o', alpha=0.5)
plt.title('Traffic Hotspots (Clusters)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
# Define the number of clusters (you can experiment with this number)
kmeans = KMeans(n_clusters=5, random_state=42)

# Fit the K-Means model
kmeans.fit(features_scaled)

# Get the cluster labels
df1['Cluster'] = kmeans.labels_

# Check the distribution of clusters
print(df1['Cluster'].value_counts())
```

```python
# Use the Elbow method to determine the optimal number of clusters
wcss = []  # Within-cluster sum of squares

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(features_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow curve
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Plot the clusters (just using Latitude and Longitude for visualization)
plt.figure(figsize=(8, 6))
plt.scatter(df1['Longitude'], df1['Latitude'], c=df1['Cluster'], cmap='viridis', marker='o', alpha=0.5)
plt.title('Traffic Hotspots (Clusters)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Get the cluster centers (in the scaled space)
centers = kmeans.cluster_centers_

# Convert back to original scale
centers = scaler.inverse_transform(centers)

# Plot the clusters with their centers
plt.figure(figsize=(8, 6))
plt.scatter(df1['Longitude'], df1['Latitude'], c=df1['Cluster'], cmap='viridis', marker='o', alpha=0.5)
plt.scatter(centers[:, 1], centers[:, 0], c='red', marker='x', s=200, label='Cluster Centers')
plt.title('Traffic Hotspots with Cluster Centers')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Cluster')
plt.legend()
plt.show()
```

```python
def predict_future_traffic_hotspot():
    print("Please enter the future accident details:")

    # Input new data for future prediction (e.g., new traffic accident)
    latitude = float(input("Latitude: "))
    longitude = float(input("Longitude: "))
    no_of_vehicles = int(input("Number of Vehicles Involved: "))
    
    # Input the other relevant features (severity and road types)
    
    severity_fatal = int(input("Severity Fatal (0/1): "))
    severity_grievous = int(input("Severity Grievous Injury (0/1): "))
    
    # Road type inputs
    road_type_nh = int(input("Road Type NH (0/1): "))
    road_type_residential = int(input("Road Type Residential Street (0/1): "))
    
    # Weather condition inputs (relevant weather conditions for prediction)
    weather_clear = int(input("Weather Clear (0/1): "))
    weather_cloudy = int(input("Weather Cloudy (0/1): "))
    weather_heavy_rain = int(input("Weather Heavy Rain (0/1): "))
    weather_light_rain = int(input("Weather Light Rain (0/1): "))
    
    # Prepare the new data point (the features for prediction)
    new_input = np.array([[latitude, longitude, no_of_vehicles,severity_fatal, 
                           severity_grievous, road_type_nh, road_type_residential, 
                           weather_clear, weather_cloudy, weather_heavy_rain, weather_light_rain]])

    # Ensure the scaler was fitted with the same feature set during training
    new_input_scaled = scaler.transform(new_input)

    # Predict the cluster for the new traffic accident using the KMeans model
    predicted_cluster = kmeans.predict(new_input_scaled)

    # Output the prediction
    print(f"\nThe predicted cluster for the new traffic accident is Cluster {predicted_cluster[0]}.")

# Run the function for future predictions
predict_future_traffic_hotspot()
```

```python
import folium

def plot_all_accident_spots(df):
    # Create a base map centered around the mean of all latitudes and longitudes
    mean_lat = df['Latitude'].mean()
    mean_lon = df['Longitude'].mean()
    
    traffic_map = folium.Map(location=[mean_lat, mean_lon], zoom_start=12)
    
    # Define a color palette for the clusters (e.g., different colors for each cluster)
    colors = ['blue', 'green', 'red', 'purple', 'orange', 'darkblue', 'lightred', 'darkgreen', 'pink']
    
    # Iterate over each row in the dataset to plot the accident locations
    for index, row in df.iterrows():
        # Ensure the cluster is an integer before using it as an index
        cluster_idx = int(row['Cluster'])  # Convert the cluster to an integer index
        cluster_color = colors[cluster_idx % len(colors)]  # This makes sure we don't exceed the color list
        
        # Add a marker for each accident with the appropriate cluster color and popup info
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"Cluster: {cluster_idx}<br>No. of Vehicles: {row['Noofvehicle_involved']}<br>Severity ACHARI: {row['Severity_ACHARI']}",
            icon=folium.Icon(color=cluster_color, icon="info-sign")
        ).add_to(traffic_map)
    
    # Save the map to an HTML file
    traffic_map.save("all_accident_hotspots_map.html")
    print("Map saved as 'all_accident_hotspots_map.html'")

# Example usage: plot all accident spots from df1
plot_all_accident_spots(df1)

```
